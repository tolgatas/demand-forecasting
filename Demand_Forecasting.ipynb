# Demand Forecasting

This notebook demonstrates how to forecast demand using ARIMA on synthetic daily sales data.

```python
import pandas as pd, matplotlib.pyplot as plt
from pathlib import Path
from statsmodels.tsa.arima.model import ARIMA

data_path = Path('../data/sample_demand.csv')
df = pd.read_csv(data_path, parse_dates=['date'])
series = df.set_index('date')['demand']
series.plot(title='Daily Demand'); plt.show()
series.head()
```

```python
# Fit ARIMA model
model = ARIMA(series, order=(2,1,2))
fit = model.fit()
forecast = fit.forecast(steps=30)
print(forecast.head())
```
